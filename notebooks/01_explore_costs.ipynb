# Exploración de Costos Operativos

Análisis de los seed CSV: viajes, gastos, rentabilidad por ruta y vehículo.

```python
import pandas as pd

trips    = pd.read_csv('../data/seed_trips.csv')
expenses = pd.read_csv('../data/seed_trip_expenses.csv')

print(trips.shape, expenses.shape)
trips.head()
```

```python
# Rentabilidad por viaje
merged = trips.merge(expenses.groupby('trip_id')['amount'].sum().reset_index(), on='trip_id', how='left')
merged['profit'] = merged['revenue'] - merged['amount'].fillna(0)
merged['margin'] = (merged['profit'] / merged['revenue'] * 100).round(1)
merged[['trip_id', 'origin', 'destination', 'revenue', 'amount', 'profit', 'margin']].head(10)
```
